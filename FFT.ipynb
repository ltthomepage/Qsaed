```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def convert_to_diffraction_pattern(image_path):
    # 读取图片并转换为灰度图像
    img = Image.open(image_path).convert('L')
    img_array = np.array(img)
    
    # 对灰度图像进行二维傅里叶变换
    img_fft = np.fft.fft2(img_array)
    
    # 对变换结果进行零频率中心化
    img_fft_shifted = np.fft.fftshift(img_fft)
    
    # 计算傅里叶变换的幅度谱
    img_fft_mag = np.abs(img_fft_shifted)
    
    # 对幅度谱进行对数变换，增强对比度
    img_fft_mag_log = np.log1p(img_fft_mag)
    
    # 显示衍射斑点图像
    plt.imshow(img_fft_mag_log, cmap='gray')
    plt.axis('off')
    plt.title('Diffraction Pattern')
    plt.show()

# 调用函数将图像转换为衍射斑点图像
image_path = "1.png"
convert_to_diffraction_pattern(image_path)
```
